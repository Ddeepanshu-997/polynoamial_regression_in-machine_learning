```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
file_path = r'D:\c++ programming\Machine Learning-A-Z-Codes-Datasets\Machine Learning A-Z (Codes and Datasets)\Part 2 - Regression\Section 6 - Polynomial Regression\Python\Position_Salaries.csv'
dataset=pd.read_csv(file_path)
print(dataset)
x=dataset.iloc[:,1:2].values
y=dataset.iloc[:,2].values
print(x)
print(y)

#no categerocal data , no splitting of sets , and no feature scaling

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor2=LinearRegression()
p=regressor.fit(x,y)

y_predict=p.predict(x)

plt.scatter(x,y)
plt.plot(x,y_predict)
plt.xlabel('posts')
plt.ylabel('salary')
plt.title('post vs salary')
plt.show()


from sklearn.preprocessing import PolynomialFeatures
polyreg=PolynomialFeatures(degree=4)
x_poly=polyreg.fit_transform(x)
q=regressor2.fit(x_poly,y)

ypoly_predict=q.predict(x_poly)

plt.scatter(x,y)
plt.plot(x,ypoly_predict)
plt.xlabel('posts')
plt.ylabel('salary')
plt.title('post vs salary')
plt.show()

plt.scatter(x,y)
plt.plot(x_poly,ypoly_predict)
plt.xlabel('posts')
plt.ylabel('salary')
plt.title('post vs salary')
plt.show()


##In summary, both approaches are valid, but using the original x values is more intuitive and easier to understand in a visualization context. 
# However, if you want to see the exact polynomial curve that the model has fitted, you can use x_poly on the x-axis for plotting.





```
